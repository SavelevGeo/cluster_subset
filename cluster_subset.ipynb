```python
from typing import Iterator, TypeVar, cast
import numpy as np
import sklearn.cluster
from IPython.display import display


def get_subsets_by_cluster_notnull(df: pd.DataFrame) -> np.ndarray:
    db = sklearn.cluster.DBSCAN(eps=0.9)
    db.fit(df.notnull())
    return db.labels_


T = TypeVar("T", bound=pd.DataFrame)


def get_subsets_without_empty_columns(
    df: T, subset_column_name: str = "subset"
) -> Iterator[T]:
    for _, subset in df.groupby(subset_column_name):
        not_empty_columns = subset.dropna(how="all", axis="columns").columns
        yield cast(T, subset[not_empty_columns])

cities = pd.read_excel("cities.xlsx").set_index("City")
cities['subset'] = get_subsets_by_cluster_notnull(cities)
display(cities)
display(cities['subset'].value_counts())

for subset in get_subsets_without_empty_columns(cities):
    display(subset)
```
