```python
from typing import Iterator
import pandas as pd
import sklearn.cluster
from IPython.display import display

def subsets_by_cluster(df: pd.DataFrame) -> Iterator[pd.DataFrame]:
    db = sklearn.cluster.DBSCAN(min_samples=3)
    db.fit(df.notnull())
    df["subset"] = db.labels_

    for _, subset in df.groupby("subset"):
        not_empty_columns = subset.dropna(how="all", axis="columns").columns
        yield subset[not_empty_columns]

cities = pd.read_excel("cities.xlsx").set_index("City")
display(cities)
for subset in subsets_by_cluster(cities):
    display(subset)
```
